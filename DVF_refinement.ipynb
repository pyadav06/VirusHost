```python
 import os
 import pandas as pd  
 import numpy as np
 from pathlib import Path
 dir= '/content/test/'
 newdir='/content/test-refined/'
 os.mkdir(newdir)
 directories = os.listdir(dir)
 y = sorted(directories)
 print(y)

#k= []
for files in y:
  print (dir+files)
  a = os.listdir(dir+files)
  os.mkdir(newdir+files)
  #print(a)
  print (dir+files+'/'+ a[0])
  k = dir+files+'/'+ a[0]
  f = pd.read_csv (k, sep='\t')
  #print (f)
  df = pd.DataFrame(f)
  #print (df)
  refined = df[df['pvalue']< 0.05]
  print(refined)
  refined.to_csv(newdir+files+'/'+ a[0], index=False)   
```

```python
dir2 = '/content/test-refined/'
directories = os.listdir(dir2)
y = sorted(directories)
#print(y)

dict1= {}
for files in y:
  #print (dir2+files)
   a = os.listdir(dir2+files)
  #os.mkdir(newdir+files)
  #print(a)
  #print (dir+files+'/'+ a[0])
   m = dir2+files+'/'+ a[0]
   f = pd.read_csv (m, sep='\t')
   f = f.reset_index()
   #print(f)
   viral_MAG = len(f)
   #print(viral_contig)
   dict1[files] = viral_MAG
  
print(dict1)
result = pd.DataFrame(list(dict1.keys()), columns= ['viral_MAG'])
result['count']=result['viral_MAG'].apply(lambda x:  dict1[x] if x in dict1 else 0)
result.to_csv('/content/DVF-refined-output', index=False)    (#add .csv extension when download the file)


  

```
