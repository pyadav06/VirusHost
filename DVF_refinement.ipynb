```python
 import os
 import pandas as pd  
 import numpy as np
 from pathlib import Path
 dir= '/content/B73-DVF-output/'
 newdir='/content/B73-DVF-refined/'
 os.mkdir(newdir)
 directories = os.listdir(dir)
 y = sorted(directories)
 print(y)

#k= []
for files in y:
  print (dir+files)
  a = os.listdir(dir+files)
  os.mkdir(newdir+files)
  #print(a)
  print(dir+files+'/'+ a[0])
  k = dir+files+'/'+ a[0]
  f = pd.read_csv(k, sep='\t')
  #print (f)
  df = pd.DataFrame(f)
  #print (df)
  refined = df[df['pvalue']< 0.05]
  print(refined)
  refined.to_csv(newdir+files+'/'+ a[0], index=False)   
```
