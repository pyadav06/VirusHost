```python
import os
from Bio import SeqIO
import pandas as pd

def get_df(dir):
      a=['A','T','G','C']
      lst=[]
      s=""
      for i in range(0,4):
        s=""
        s=s+a[i]
        for j in range(0,4):
            t=s+a[j]
            for k in range(0,4):
                u=t+a[k]
                for l in range(0,4):
                    v=u+a[l]
                    lst.append(v)
      
      df = pd.DataFrame (lst, columns = ['Combination'])

      directories = os.listdir(dir)
      
      # This would print all the files and directories
      for file in directories:
          filename=str(file).replace(".fa","")
          window_size = 4
          dict1={}
          c=0
          filepath=dir+"/"+str(filename)+".fa"
          boolean=os.path.exists(filepath)
          if boolean==False:
            continue
          for seq_record in SeqIO.parse(filepath, "fasta"):
          #   print(seq_record.id)
              x=seq_record.seq
              for i in range(len(x) - window_size + 1):
                  a=str(x[i: i + window_size])
                  c=c+1
                  if a in dict1:
                      dict1[a]+=1
                  else:
                      dict1[a]=1

          #   print(len(seq_record))
          for k,v in dict1.items():
            dict1[k]=v/c
          df[filename]=df['Combination'].apply(lambda x:  dict1[x] if x in dict1 else 0)
      return df 
```

```python
Bacteria=get_df("/content/B73-rhizosphere/Bacteria")
Bacteria
```

```python
#Change the file extension name in the main script to run virus files (example, virus.fasta)
Virus=get_df("/content/B73-rhizosphere/Virus")
Virus 
```

```python
from scipy.spatial import distance
virus_list=list(Virus.columns[1:])
bacteria_list=list(Bacteria.columns[1:])
lst_corr=[]

for i in virus_list:
  for j in bacteria_list:
    bacteria_series=Bacteria[j]
    virus_series=Virus[i]
    dst = distance.euclidean(bacteria_series, virus_series)
    lst_corr.append((j,i,dst))
result=pd.DataFrame(lst_corr,columns = ['Bacteria','Virus','Euclidean-distance'])
result   
```
